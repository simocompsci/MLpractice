```python
from sklearn.neighbors import KNeighborsClassifier

# données météo
X = [
    [30,70,10],
    [25,90,5],
    [28,85,7],
    [35,40,12],
    [27,80,6]
]

y = ["Non","Oui","Oui","Non","Oui"]

# modèle KNN
knn = KNeighborsClassifier(n_neighbors=2,metric='manhattan')
knn.fit(X, y)

# nouvelle météo
prediction = knn.predict([[31,72,13]])

print("Pluie prévue :", prediction)
```
