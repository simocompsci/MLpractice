```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler

X = np.array([[5], [10], [15], [20], [100]])

# MinMaxScaler
minmax = MinMaxScaler()
X_minmax = minmax.fit_transform(X)

print("\n--- MinMaxScaler ---")
print("Min :", minmax.data_min_)
print("Max :", minmax.data_max_)
print("Données normalisées :")
print(X_minmax)
# StandardScaler
standard = StandardScaler()
X_standard = standard.fit_transform(X)

print("\n--- StandardScaler ---")
print("Moyenne :", standard.mean_)
print("Variance :", standard.var_)
print("Écart-type :", np.sqrt(standard.var_))
print("Données standardisées :")
print(X_standard)
```
